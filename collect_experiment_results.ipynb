```python
import os 
import glob
import pickle

import pandas as pd 
```

```python
def read_pickle( fname ) : 
    with open( fname, "rb" ) as f_in : 
        return pickle.load( f_in )
```

```python
import datetime as dt 
dt.datetime.fromtimestamp( os.stat( in_path ).st_mtime )
```

```python
all_dics = []
for in_path in glob.glob( "experiment_results/exp*.pkl") :
    
    dic = read_pickle( in_path )    
    del dic["per_epoch"]
    dic["time"] = dt.datetime.fromtimestamp( os.stat( in_path ).st_mtime )
    all_dics.append( dic )
    
df = pd.DataFrame( all_dics ).sort_values( "best_valid_accy", ascending=False) 
df = df.set_index( 'md5_dig')[["best_valid_accy",  "best_valid_epoch", "netw_arch_name",
                                "learning_rate", "batch_size", "keep_prob", "time"]]
#df = pd.DataFrame( all_dics ).sort_values( "time", ascending=False) 
df
```
